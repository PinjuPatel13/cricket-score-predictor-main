```python
import pandas as pd

# Assuming df is your DataFrame

# Convert 'runs' column to numeric
df['runs'] = pd.to_numeric(df['runs'], errors='coerce')

# Calculate the cumulative sum within each group
df['current_score'] = df.groupby('match_id')['runs'].cumsum()

# If you want to fill NaN values with 0 in the 'current_score' column
df['current_score'] = df['current_score'].fillna(0)

# Print the DataFrame to check the result
print(df)
```

```python
df
```

```python
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
df
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
df
```

```python
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
df
```

```python
import pandas as pd

# Assuming df is your DataFrame

# Convert 'player_dismissed' to 0 or 1
df['player_dismissed'] = df['player_dismissed'].apply(lambda x: 0 if x == '0' else 1)

# Convert 'player_dismissed' to integer
df['player_dismissed'] = df['player_dismissed'].astype('int')

# Calculate the cumulative sum of 'player_dismissed' within each group
df['player_dismissed'] = df.groupby('match_id')['player_dismissed'].cumsum()

# Calculate the number of wickets left
df['wickets_left'] = 10 - df['player_dismissed']

# Print the DataFrame to check the result
print(df)
```

```python
df
```

```python
df['crr'] = (df['current_score']*6)/df['balls_bowled']
```

```python
df
```

```python
import pandas as pd

# Assuming df is your DataFrame

# Convert 'runs' column to numeric, coerce errors to NaN
df['runs'] = pd.to_numeric(df['runs'], errors='coerce')

# Group by 'match_id'
groups = df.groupby('match_id')

# Get unique match_ids
match_ids = df['match_id'].unique()

# Create an empty list to store the last five runs for each match
last_five = []

# Iterate over each match_id
for id in match_ids:
    # Get the group for the current match_id
    group = groups.get_group(id)
    
    # Calculate the rolling sum of 'runs' for the last 30 balls and add to last_five
    last_five.extend(group['runs'].rolling(window=30, min_periods=1).sum().values.tolist())

# Print the last five runs for each match
print(last_five)
```

```python
df['last_five'] = last_five
```

```python
df[['match_id', 'batting_team', 'bowling_team', 'ball', 'runs', 'player_dismissed', 'city', 'current_score', 'over', 'ball_no', 'balls_bowled', 'balls_left', 'wickets_left', 'crr', 'last_five']]
```

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
final_df=final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','runs_x']]
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.isnull().sum()
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.sample(2)
```

```python
X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
trf = ColumnTransformer([
    ('trf', OneHotEncoder(drop='first'), ['batting_team', 'bowling_team', 'city'])
], remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('trf', trf),  # Apply the ColumnTransformer first
    ('scaler', StandardScaler(with_mean=False)),  # Scale the numeric columns
    ('regressor', XGBRegressor(n_estimators=1000, learning_rate=0.2, max_depth=12, random_state=1))
])
```

```python
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
print("R2 Score:", r2_score(y_test, y_pred))
print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
```

```python
pickle.dump(pipe,open('pipe.pkl','wb'))
```
